```python
# Code to calculate internal temperatures, SPC and ISP of a Sea-Dart ramjet and Harpoon turbojet, post boost phase

# Packages imported
import numpy as np
import pandas as pd

# Standard variables
R = 0.285e3  # kJ/kgK - gas constant
gravity = 9.81  # m/s
Cp_air = 1.005  # kJ/kgK
Cp_gas = 1.147  # kJ/kgK
Y_air = 1.4
Y_gas = 1.33
Hc = 43  # MJ/kg

density_modifier = 1.2250  # from IAS table
# IAS_online = pd.read_html('https://www.engineeringtoolbox.com/international-standard-atmosphere-d_985.html')
IAS_stand_atm = pd.read_excel(r'C:\Users\Liam\Documents\Masters\Resources\ISA_table.xlsx')
```

### 1. Question One {{ref_bray}} ###

A Sea-Dart ramjet missile has the following characteristics at the start of its cruise phase:

- Post-boost weight $(w)$ = 500 kg, 
- L/D ratio $(\frac{L}{D})$ = 0.5, 
- Cruise speed $(M_a)$ = mach 2.5. 

The power-plant uses a single oblique and normal shock under cruise conditions and displays the following characteristics: 

- Intake area $(A)$ = 0.01$m^2$, 
- Pressure drop across the combuster = 5%,
- $\gamma$ = 1.33, 
- $R$ = 0.285 kJ/kgK.

Using these variables, the standard variables above and Figures 1.1 and 1.2 below, calculate metrics 1-3 listed below for initial cruise conditions; clearly stating all assumptions made:
1. Combustion temperature, 
2. Specific fuel consumption,
3. Specific impulse of the engine.

```python
## QUESTION 1

# Assumptions
# lift   = weight X gravity - body in equilibrium
# thrust = drag = body at constant speed
# exit mass-flow = input mass-flow
# Po6 = Po3 - perfect nozzle (normally around 0.9 - 0.95)
# P6 = Pa
# To4 = To6

# Missile variables
mach_missile = 2.5  # mach
weight_missile = 500  # kg
recovery_value = 0.65  # from graph for 2.5 mach
lift_to_drag = 0.5
pressure_recovery = 0.95  # 5% drop
area_intake = 0.01  # m^2
nozzle_loss = 1

# Altitude inputs
altitude = 0e3  # meters
# altitude = (float(input('Enter the cruising altitude of the SeaDart in km '))*1000)

temperature_a = IAS_stand_atm['Temp'].where(IAS_stand_atm['Alt'] == altitude)
temperature_a = temperature_a.nlargest(1)
temperature_a = float(temperature_a.to_numpy())
speed_of_sound_a = IAS_stand_atm['a'].where(IAS_stand_atm['Alt'] == altitude)  # look up value at alt
speed_of_sound_a = speed_of_sound_a.nlargest(1)  # select the value
speed_of_sound_a = float(speed_of_sound_a.to_numpy())  # convert to array and then float
pressure_a = IAS_stand_atm['Px'].where(IAS_stand_atm['Alt'] == altitude)
pressure_a = pressure_a.nlargest(1)
pressure_a = float(pressure_a.to_numpy())
density_a = IAS_stand_atm['Dens'].where(IAS_stand_atm['Alt'] == altitude)
density_a = density_a.nlargest(1)
density_a = (float(density_a.to_numpy())) * density_modifier

# Missile thrust (F) in Newtons
lift = weight_missile * gravity
drag = lift / lift_to_drag
thrust = drag

# Local velocity at front face of the missile (Ua) in m/s
velocity_a = mach_missile * speed_of_sound_a

# Mass flow into the intakes and out of the nozzle in kg/s
mass_flow_a = density_a * area_intake * velocity_a
mass_flow_e = mass_flow_a

# Velocity at missile exit (Ue) in m/s
velocity_exit = (thrust / mass_flow_a) + velocity_a

# Local total temperature at front face of the missile (Toa) in kelvin
temperature_total_a = (velocity_a ** 2 / (2 * Cp_air * 1000)) + temperature_a

# Exit Mach (Me)
# Po/P = (To/T)**(Y_air/(Y_air-1))
# Po/P = (1+ (y-1/2)*m^2)^(Y/Y-1)
pressure_total_a = pressure_a * ((temperature_total_a / temperature_a) ** (Y_air / (Y_air - 1)))
pressure_total_1 = pressure_total_a
pressure_total_2 = pressure_total_a * recovery_value
pressure_total_3 = pressure_total_2 * pressure_recovery
pressure_total_6 = pressure_total_3 * nozzle_loss
pressure_6 = pressure_a

mach_exit = np.sqrt((2 * ((((pressure_total_6 / pressure_6) ** ((Y_gas - 1) / Y_gas))) - 1)) / (Y_gas - 1))

# Exit Temperature (T6) in Kelvin
speed_of_sound_at_exit = velocity_exit / mach_exit
temperature_6 = speed_of_sound_at_exit ** 2 / (Y_gas * R)

# Temperature in combustion chamber (T04) in Kelvin
# To6/T6 = 1 + ((Y - 1) / 2) * Me^2
temperature_total_6 = temperature_6 * (1 + ((Y_gas - 1) / 2) * mach_exit ** 2)
temperature_total_4 = temperature_total_6

# User to input value from fuel to air graph (1/f)
air_fuel_ratio_graph = 12
# air_fuel_ratio = float(input('Enter the air/ fuel ratio for a combustion temperature of ' + str(int(temperature_total_4)) + ' kelvin for the SeaDart'))
fuel_air_ratio = 1 / air_fuel_ratio_graph

# SFC and ISP (in kg/s/N and N)
m_dot_f = mass_flow_a * fuel_air_ratio
specific_fuel_consumption = m_dot_f / thrust
specific_impulse = 1 / specific_fuel_consumption

# Print
#print()
#print('The combustion temperature for the SeaDart is ' + str(round(temperature_total_4 ,3)) + ' Kelvin')
#print('The SFC for the SeaDart is ' + str(round (specific_fuel_consumption, 8)) + ' kg/s/N')
#print('The ISP for the SeaDart is ' + str(round(specific_impulse, 3)) + ' N')
```

### 2. Question One Answer ####

#### 2.1 Assumptions ####

To answer question one the following <b>assumptions</b> have been made:

- The missile is nether climbing or falling, therefore lift = weight,
- The missile is traveling at a constant velocity, therefore thrust = drag,
- Missile is fully expanded and entry mass flow rate = exit mass flow rate,
- Nozzle is perfect, and therefore P$_o$6 = P$_o$3,
- Exhaust is fully expanded, therefore P6 = P$_a$,
- Total temperature loss through he nozzle is assumed to be zero, therefore T$_o$3 (combustion chamber temperature) = T$_o$6,
- Sea Dart has a maximum engagement altitude of 18.5 km {{ref_jane_sd}}. However this simulation will assume a sea skimming profile and, therefore, a cruising altitude of {{int(altitude/1000)}} km.

#### 2.5 Temperature $(T_6)$ and Mach $(M_e)$ at Exit

```python
# Code to calculate internal temperatures, SPC and ISP of a Sea-Dart ramjet and Harpoon turbojet, post boost phase

# Packages imported
import numpy as np
import pandas as pd

# Standard variables
R = 0.285e3  # kJ/kgK - gas constant
gravity = 9.81  # m/s
Cp_air = 1.005  # kJ/kgK
Cp_gas = 1.147  # kJ/kgK
Y_air = 1.4
Y_gas = 1.33
Hc = 43  # MJ/kg

density_modifier = 1.2250  # from IAS table
# IAS_online = pd.read_html('https://www.engineeringtoolbox.com/international-standard-atmosphere-d_985.html')
IAS_stand_atm = pd.read_excel(r'C:\Users\Liam\Documents\Masters\Resources\ISA_table.xlsx')

############################################
## QUESTION 1

# Assumptions
# lift   = weight X gravity - body in equilibrium
# thrust = drag = body at constant speed
# exit mass-flow = input mass-flow
# Po6 = Po3 - perfect nozzle (normally around 0.9 - 0.95)
# P6 = Pa
# To4 = To6

# Missile variables
mach_missile = 2.5  # mach
weight_missile = 500  # kg
recovery_value = 0.65  # from graph for 2.5 mach
lift_to_drag = 0.5
pressure_recovery = 0.95  # 5% drop
area_intake = 0.01  # m^2
nozzle_loss = 1

# Altitude inputs
altitude = 0e3  # meters
# altitude = (float(input('Enter the cruising altitude of the SeaDart in km '))*1000)

temperature_a = IAS_stand_atm['Temp'].where(IAS_stand_atm['Alt'] == altitude)
temperature_a = temperature_a.nlargest(1)
temperature_a = float(temperature_a.to_numpy())
speed_of_sound_a = IAS_stand_atm['a'].where(IAS_stand_atm['Alt'] == altitude)  # look up value at alt
speed_of_sound_a = speed_of_sound_a.nlargest(1)  # select the value
speed_of_sound_a = float(speed_of_sound_a.to_numpy())  # convert to array and then float
pressure_a = IAS_stand_atm['Px'].where(IAS_stand_atm['Alt'] == altitude)
pressure_a = pressure_a.nlargest(1)
pressure_a = float(pressure_a.to_numpy())
density_a = IAS_stand_atm['Dens'].where(IAS_stand_atm['Alt'] == altitude)
density_a = density_a.nlargest(1)
density_a = (float(density_a.to_numpy())) * density_modifier

# Missile thrust (F) in Newtons
lift = weight_missile * gravity
drag = lift / lift_to_drag
thrust = drag

# Local velocity at front face of the missile (Ua) in m/s
velocity_a = mach_missile * speed_of_sound_a

# Mass flow into the intakes and out of the nozzle in kg/s
mass_flow_a = density_a * area_intake * velocity_a
mass_flow_e = mass_flow_a

# Velocity at missile exit (Ue) in m/s
velocity_exit = (thrust / mass_flow_a) + velocity_a

# Local total temperature at front face of the missile (Toa) in kelvin
temperature_total_a = (velocity_a ** 2 / (2 * Cp_air * 1000)) + temperature_a

# Exit Mach (Me)
# Po/P = (To/T)**(Y_air/(Y_air-1))
# Po/P = (1+ (y-1/2)*m^2)^(Y/Y-1)
pressure_total_a = pressure_a * ((temperature_total_a / temperature_a) ** (Y_air / (Y_air - 1)))
pressure_total_1 = pressure_total_a
pressure_total_2 = pressure_total_a * recovery_value
pressure_total_3 = pressure_total_2 * pressure_recovery
pressure_total_6 = pressure_total_3 * nozzle_loss
pressure_6 = pressure_a

mach_exit = np.sqrt((2 * ((((pressure_total_6 / pressure_6) ** ((Y_gas - 1) / Y_gas))) - 1)) / (Y_gas - 1))

# Exit Temperature (T6) in Kelvin
speed_of_sound_at_exit = velocity_exit / mach_exit
temperature_6 = speed_of_sound_at_exit ** 2 / (Y_gas * R)

# Temperature in combustion chamber (T04) in Kelvin
# To6/T6 = 1 + ((Y - 1) / 2) * Me^2
temperature_total_6 = temperature_6 * (1 + ((Y_gas - 1) / 2) * mach_exit ** 2)
temperature_total_4 = temperature_total_6

# User to input value from fuel to air graph (1/f)
air_fuel_ratio_graph = 12
# air_fuel_ratio = float(input('Enter the air/ fuel ratio for a combustion temperature of ' + str(int(temperature_total_4)) + ' kelvin for the SeaDart'))
fuel_air_ratio = 1 / air_fuel_ratio_graph

# SFC and ISP (in kg/s/N and N)
m_dot_f = mass_flow_a * fuel_air_ratio
specific_fuel_consumption = m_dot_f / thrust
specific_impulse = 1 / specific_fuel_consumption

# Print
print('RESULTS')
print()
print()
print('The combustion temperature for the SeaDart is ' + str(round(temperature_total_4 ,3)) + ' K')
print('The SFC for the SeaDart is ' + str(round (specific_fuel_consumption, 8)) + ' kg/Ns')
print('The ISP for the SeaDart is ' + str(round(specific_impulse, 3)) + ' Ns/kg')

#########################################################
## QUESTION 2

# Assumptions
# L = D-
# T = W-
# fully expanded-
# nozzle perfect, no pressure lost
# total temp 2 = total temp a - no loss in intake
# assume complete combustion of fuel
# to5 = to9
# compressor work = turbine work - assumed 100% efficient

# Missile variables
weight_missile = 600  # kg (mass)
lift_to_drag = 2
mach_missile = 0.883  # mach (sub sonic)

area_intake = 0.05  # m^2
pressure_recovery_intake = 0.95

pressure_ratio_1_compressor = 1.2 / 1
pressure_ratio_2_compressor = 4 / 1
isentropic_efficiency_compressor = 0.85

pressure_recovery_combuster = 0.95

pressure_ratio_turbine = 4 / 1
temperature_max_entry_turbine = 1200  # kelvin
isentropic_efficiency_turbine = 0.95

diameter_nozzle = 150e-3  # m
pressure_recovery_nozzle = 1

# Local conditions
altitude = 0  # skimming at sea-level
# print('')
# print('')
# altitude = (float(input('Enter the cruising altitude of the Harpoon in km (0 is recomended due to its' sea-skimming profile) '))*1000)

temperature_a = IAS_stand_atm['Temp'].where(IAS_stand_atm['Alt'] == altitude)
temperature_a = temperature_a.nlargest(1)
temperature_a = float(temperature_a.to_numpy())
speed_of_sound_a = IAS_stand_atm['a'].where(IAS_stand_atm['Alt'] == altitude)  # look up value at alt
speed_of_sound_a = speed_of_sound_a.nlargest(1)  # select the value
speed_of_sound_a = float(speed_of_sound_a.to_numpy())  # convert to array and then float
pressure_a = IAS_stand_atm['Px'].where(IAS_stand_atm['Alt'] == altitude)
pressure_a = pressure_a.nlargest(1)
pressure_a = float(pressure_a.to_numpy())
density_a = IAS_stand_atm['Dens'].where(IAS_stand_atm['Alt'] == altitude)
density_a = density_a.nlargest(1)
density_a = (float(density_a.to_numpy())) * density_modifier

# Thrust
lift = weight_missile * gravity
drag = lift / lift_to_drag
thrust = drag

# Local velocity at front face of the missile (Ua) in m/s
velocity_a = mach_missile * speed_of_sound_a

# Local total temperature at front face of the missile (Toa) in kelvin
temperature_total_a = (velocity_a ** 2 / (2 * Cp_air * 1000)) + temperature_a

# pressures and temps through to exit mach
# a - ambient
# 1 - entrance to intake
# 2 - after intake, entry to compressor
# 2_mid - mid compressor after first stage
# 3 - entry to combuster
# 4 - entry to turbine
# 5 - exit of turbine
# 6 - entry to nozzle
# 9 - exit of nozzle

# Po/P = (To/T)**(Y_air/(Y_air-1))
# Po/P = (1+ (y-1/2)*m^2)^(Y/Y-1)
# ambient = 1
pressure_total_a = pressure_a * ((temperature_total_a / temperature_a) ** (Y_air / (Y_air - 1)))

# 2 - at compressor (after intake)
pressure_total_2 = pressure_total_a * pressure_recovery_intake
temperature_total_2 = temperature_total_a  # assume no temp loss (total)

# 2_mid - mid compressor
pressure_total_2_mid = pressure_total_2 * pressure_ratio_1_compressor

# 3 - at combustion chamber (after compressor)
pressure_total_3 = pressure_total_2_mid * pressure_ratio_2_compressor
temperature_total_3_isentropic = temperature_total_2 * ((pressure_total_3 / pressure_total_2) ** ((Y_air - 1) / Y_air))
temperature_total_3 = ((temperature_total_3_isentropic - temperature_total_2) / isentropic_efficiency_compressor) + temperature_total_2

# 4 - entry to turbine (post combustion)
pressure_total_4 = pressure_total_3 * pressure_recovery_combuster

# 5 - post turbine
pressure_total_5 = pressure_total_4 / pressure_ratio_turbine  # at nozzle

# 9 - exit start
pressure_total_9 = pressure_total_5 * pressure_recovery_nozzle
pressure_9 = pressure_a  # assume fully expanded

# mass flow into intake
mass_flow_a = density_a * area_intake * velocity_a
# Velocity at missile exit (Ue) in m/s
velocity_exit = (thrust / mass_flow_a) + velocity_a
# exit mach and speed of sound
mach_exit = np.sqrt((2 * ((((pressure_total_9 / pressure_9) ** ((Y_gas - 1) / Y_gas))) - 1)) / (Y_gas - 1))
speed_of_sound_at_exit = velocity_exit / mach_exit

temperature_9 = speed_of_sound_at_exit ** 2 / (Y_gas * R)
temperature_total_9 = temperature_9 * (1 + ((Y_gas - 1) / 2) * mach_exit ** 2)

# Assumed
temperature_total_5 = temperature_total_9

# Mass flow  out of the nozzle
mass_flow_e = mass_flow_a

# temp at combustion
# calculate the compressor work
compressor_work_done = mass_flow_a * Cp_air * 1000 * (temperature_total_3 - temperature_total_2)
# Turbine work done
turbine_work_done = compressor_work_done  # assumed 100% efficient

temperature_total_4 = (turbine_work_done / (mass_flow_a * Cp_gas * 1000)) + temperature_total_5
combustion_temperature = temperature_total_4

temperature_rise_in_combustion_chamber = temperature_total_4 - temperature_total_3

print()
print('Using method one to calculate combustion temperature in the Harpoon')
print('Temperature rise in combustion chamber = ' + str(round(temperature_rise_in_combustion_chamber, 3)) + ' kelvin')

# Air fuel ratio
# air_fuel_ratio = float(input('enter the air/ fuel ratio for a combustion temperature rise of ' + str(int(temperature_rise_in_combustion_chamber)) +' kelvin for the Harpoon'))
#0.0078 from graph
fuel_air_ratio_graph = 0.0078
fuel_air_ratio = fuel_air_ratio_graph

# SFC and ISP (in kg/s/N and N)
m_dot_f = mass_flow_a * fuel_air_ratio


specific_fuel_consumption = m_dot_f / thrust
specific_impulse = 1 / specific_fuel_consumption

# Print
print('The combustion temperature for the Harpoon is ' + str(round(temperature_total_4 ,3)) + ' K')
print('The SFC for the Harpoon is ' + str(round (specific_fuel_consumption, 8)) + ' kg/Ns')
print('The ISP for the Harpoon is ' + str(round(specific_impulse, 3)) + ' Ns/kg')
print('')

# second method for working out To4

# Wt = Wc - assumption
# m_dot*Cp_gas*(d_To45) = m_dot*Cp_air*(To3-To2)
# d_To45 = (Cp_air/Cp_gas)*(To3-To2)
delta_temperature_total_4_5 = (Cp_air/Cp_gas)*(temperature_total_3 - temperature_total_2)
# d_To45 = To4*eff[1-(po5/po4)**(y-1/y)
# To4 = (DTo4-5)/eff*(1-(po5/po4)**(y-1/y))
temperature_total_4 = delta_temperature_total_4_5/ (isentropic_efficiency_turbine * (1 - (pressure_total_5/pressure_total_4)**((Y_gas-1)/Y_gas)))
#temperature rise
temperature_rise_in_combustion_chamber = temperature_total_4 - temperature_total_3

print()
print('Using method two to calculate combustion temperature in the Harpoon')
print('Temperature rise in combustion chamber = ' + str(round(temperature_rise_in_combustion_chamber, 3)) + ' kelvin')

# 0.0035 from graph
fuel_air_ratio_graph = 0.0035

fuel_air_ratio = fuel_air_ratio_graph

# SFC and ISP (in kg/s/N and N)
m_dot_f = mass_flow_a * fuel_air_ratio

specific_fuel_consumption = m_dot_f / thrust
specific_impulse = 1 / specific_fuel_consumption

# Print
print('The combustion temperature for the Harpoon is ' + str(round(temperature_total_4 ,3)) + ' K')
print('The SFC for the Harpoon is ' + str(round (specific_fuel_consumption, 8)) + ' kg/Ns')
print('The ISP for the Harpoon is ' + str(round(specific_impulse, 3)) + ' Ns/kg')

```
